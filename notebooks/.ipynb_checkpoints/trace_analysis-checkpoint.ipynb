# DeepRacer trace analysis

Analyses a dr-gym training run end-to-end from the **Tier-1 trace**
(`gym_dr.trace`, see `docs/trace-contract.md`), and cross-checks it against
**MLflow**, **TensorBoard**, and **Optuna**.

Two halves:

1. **Original `deepracer-utils` analysis, tested on our trace.** Our Tier-1
   columns deliberately match the deepracer-utils *internal* DataFrame, so its
   `AnalysisUtils` / `SimtraceStabilityAnalyzer` run on our data with only a
   thin column adapter — no S3, no log loaders.
2. **New analyses** built on the trace + MLflow + TensorBoard + Optuna that
   deepracer-utils never had (per-world, cross-trial, multimodal-ready).

Run it with the project's `analysis` dep group:

```
uv run --group analysis jupyter lab notebooks/trace_analysis.ipynb
```


````python
import sys, glob
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- locate the project + deepracer-utils, regardless of cwd ---------------
def find_project_root(start=None):
    p = Path(start or Path.cwd()).resolve()
    for cand in [p, *p.parents]:
        if (cand / "optuna.db").exists() or (cand / "artifacts").is_dir():
            return cand
    return p

PROJECT = find_project_root()
DU_CANDIDATES = [PROJECT.parent / "deepracer-utils", Path("/home/lunav0/Projects/deepracer-utils")]
DEEPRACER_UTILS = next((p for p in DU_CANDIDATES if p.exists()), None)
if DEEPRACER_UTILS:
    sys.path.insert(0, str(DEEPRACER_UTILS))
sys.path.insert(0, str(PROJECT))  # so `import gym_dr` works from notebooks/

print("project        :", PROJECT)
print("deepracer-utils:", DEEPRACER_UTILS)
````

## 1. Load the Tier-1 trace

`gym_dr.trace.load_steps` concatenates the per-episode Parquet shards each run wrote under `artifacts/<run>/trace/steps/`. We load every run that has a trace and tag rows with their run dir.

````python
from gym_dr.trace import load_steps, STEP_COLUMNS

run_dirs = sorted(p.parent.parent for p in PROJECT.glob("artifacts/*/trace/steps"))
assert run_dirs, "no runs with a trace found — enable TraceConfig(enabled=True) and train"

frames = []
for rd in run_dirs:
    d = load_steps(rd)
    if len(d):
        d = d.assign(run=rd.name)
        frames.append(d)
df = pd.concat(frames, ignore_index=True)

print("runs with trace:", [rd.name for rd in run_dirs])
print("rows:", len(df), "| schema matches contract:", list(df.columns)[:len(STEP_COLUMNS)] == STEP_COLUMNS)
print("worlds:", df["world_name"].dropna().unique().tolist())
df.head(3)
````

## 2. Per-episode summary (our own, schema-native)

A quick groupby straight off the trace — no adapter needed.

````python
g = df.groupby(["run", "world_name", "chunk_index", "episode"], dropna=False)
summary = g.agg(
    steps=("steps", "max"),
    max_progress=("progress", "max"),
    total_reward=("reward", "sum"),
    mean_speed=("speed", "mean"),
    offtrack_steps=("on_track", lambda s: int((~s).sum())),
    crashes=("is_crashed", "sum"),
    status=("episode_status", "last"),
).reset_index()
print(f"{len(summary)} episodes")
display(summary.tail(10))

fig, ax = plt.subplots(1, 2, figsize=(12, 3.5))
ax[0].plot(summary["max_progress"].values, marker="o", ms=3); ax[0].set(title="Max progress per episode", xlabel="episode #", ylabel="progress %")
ax[1].plot(summary["total_reward"].values, marker="o", ms=3, color="tab:green"); ax[1].set(title="Total reward per episode", xlabel="episode #", ylabel="reward")
for a in ax: a.grid(alpha=.3)
plt.tight_layout(); plt.show()
````

## 3. Adapter → deepracer-utils schema

deepracer-utils expects a few columns our in-process trace doesn't carry (`tstamp`, `wall_clock`, `worker`, `iteration`). We map them:

- `tstamp`/`wall_clock` ← `wall_time` (the in-process producer has no sim `/clock`; the bag→trace path fills real `sim_time` later — see the contract).
- `worker` ← 0 (single-worker), `iteration` ← `chunk_index`.

Everything else already lines up by name (`steering_angle`, `speed`, `on_track`, `progress`, `closest_waypoint`, `track_len`, `episode_status`).

````python
def to_deepracer_df(trace: pd.DataFrame) -> pd.DataFrame:
    """Adapt a Tier-1 trace to the deepracer-utils internal DataFrame schema."""
    out = trace.copy()
    clock = out["sim_time"].where(out["sim_time"].notna(), out["wall_time"])
    out["tstamp"] = clock.astype(float)
    out["wall_clock"] = out["wall_time"].astype(float)
    out["worker"] = 0
    out["iteration"] = out["chunk_index"].fillna(0).astype(int)
    out["unique_episode"] = out["episode"].astype(int)
    return out

du = to_deepracer_df(df)
du[["episode", "iteration", "worker", "tstamp", "steps", "x", "y", "speed", "progress", "episode_status"]].head(3)
````

## 4. deepracer-utils `AnalysisUtils.simulation_agg` (original code, our data)

This is the **unmodified** deepracer-utils per-episode aggregator running on our adapted trace — proof the analysis layer ports without a rewrite.

````python
try:
    from deepracer.logs import AnalysisUtils
    agg = AnalysisUtils.simulation_agg(du.copy(), firstgroup="iteration", secondgroup="episode", add_tstamp=True)
    display(agg.tail(8))
    try:
        AnalysisUtils.scatter_aggregates(agg, title="deepracer-utils scatter_aggregates (on dr-gym trace)")
        plt.show()
    except Exception as e:
        print("scatter_aggregates skipped:", e)
except Exception as e:
    print("AnalysisUtils.simulation_agg unavailable / failed:", repr(e))
````

## 5. deepracer-utils `SimtraceStabilityAnalyzer` (step-timing stability)

Per-iteration step-delta statistics. Here the clock is `wall_time` (the in-process trace has no sim clock), so `rtf≈1` by construction — once the bag→trace path supplies real `sim_time`, this yields a true real-time factor.

````python
try:
    from deepracer.logs import SimtraceStabilityAnalyzer
    stab = SimtraceStabilityAnalyzer(du.copy()).analyze()
    display(stab)
except Exception as e:
    print("stability analysis failed:", repr(e))
````

## 6. New: trajectory, coloured by speed

Track-geometry-free — straight from `x/y/speed`. Picks the longest episode in the latest run. (With per-world waypoints loaded, deepracer-utils' `PlottingUtils.plot_track` would overlay this on the track outline.)

````python
latest = df[df["run"] == df["run"].iloc[-1]]
ep_id = latest.groupby("episode")["steps"].max().idxmax()
ep = latest[latest["episode"] == ep_id]
fig, ax = plt.subplots(figsize=(6, 6))
sc = ax.scatter(ep["x"], ep["y"], c=ep["speed"], cmap="viridis", s=10)
ax.plot(ep["x"], ep["y"], lw=.3, color="grey", alpha=.5)
ax.set(title=f"Trajectory — run={ep['run'].iloc[0]} ep={ep_id} ({len(ep)} steps)", xlabel="x (m)", ylabel="y (m)")
ax.set_aspect("equal", "box"); fig.colorbar(sc, label="speed (m/s)"); plt.show()
````

## 7. New: MLflow run metrics

Pulls the `trace-smoke` experiment from the local MLflow store and plots the `dr/ep_*` metric history the mirror callback logged.

````python
import mlflow
from mlflow.tracking import MlflowClient

mlflow.set_tracking_uri(f"file:{PROJECT / 'mlruns'}")
client = MlflowClient()
exp = mlflow.get_experiment_by_name("trace-smoke")
if exp is None:
    print("no 'trace-smoke' experiment yet")
else:
    runs = mlflow.search_runs([exp.experiment_id])
    keep = [c for c in ["tags.mlflow.runName", "status", "metrics.dr/ep_reward",
                        "metrics.dr/ep_max_progress", "metrics.eval/mean_reward",
                        "params.trainer.kwargs.learning_rate"] if c in runs.columns]
    display(runs[keep])

    # Metric history for the most recent run.
    rid = runs.sort_values("start_time")["run_id"].iloc[-1]
    fig, ax = plt.subplots(1, 2, figsize=(12, 3.5))
    for key, a in zip(["dr/ep_reward", "dr/ep_max_progress"], ax):
        hist = client.get_metric_history(rid, key)
        if hist:
            xs = [h.step for h in hist]; ys = [h.value for h in hist]
            a.plot(xs, ys, marker=".", ms=4); a.set(title=key, xlabel="timestep"); a.grid(alpha=.3)
        else:
            a.set_title(f"{key} (no history)")
    plt.tight_layout(); plt.show()
````

## 8. New: TensorBoard scalars

Reads the SB3 event file directly with TensorBoard's `EventAccumulator` and plots the PPO learning curve — the same source the TB UI uses.

````python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

ev_files = sorted(PROJECT.glob("artifacts/*/tensorboard/**/events.out.tfevents.*"))
if not ev_files:
    print("no TensorBoard event files yet")
else:
    ea = EventAccumulator(str(ev_files[-1].parent)); ea.Reload()
    tags = ea.Tags().get("scalars", [])
    print("scalar tags:", tags)
    want = next((t for t in ["rollout/ep_rew_mean", "train/value_loss", "train/loss"] if t in tags), tags[0] if tags else None)
    if want:
        evs = ea.Scalars(want)
        plt.figure(figsize=(7, 3.5))
        plt.plot([e.step for e in evs], [e.value for e in evs], marker=".", ms=3)
        plt.title(f"TensorBoard: {want}"); plt.xlabel("timestep"); plt.grid(alpha=.3); plt.show()
````

## 9. New: Optuna study

The shared SQLite study that drove the trials — values + the params each trial sampled. (Param-importance needs several completed trials; guarded.)

````python
import optuna
try:
    study = optuna.load_study(study_name="trace_smoke", storage=f"sqlite:///{PROJECT / 'optuna.db'}")
    tdf = study.trials_dataframe()
    display(tdf[[c for c in tdf.columns if c in ("number", "value", "state") or c.startswith("params_")]])
    done = [t for t in study.trials if t.value is not None]
    if len(done) >= 1:
        plt.figure(figsize=(6, 3.2))
        plt.plot([t.number for t in done], [t.value for t in done], marker="o")
        plt.title("Optuna trial value (eval reward)"); plt.xlabel("trial #"); plt.ylabel("value"); plt.grid(alpha=.3); plt.show()
except Exception as e:
    print("optuna study not available:", repr(e))
````

## 10. New: off-track / object-avoidance summary

Straight off the trace's OA columns — the kind of per-episode safety accounting deepracer-utils predates. (`oa_enabled=False` here since this run is pure time-trial; the columns are still exercised.)

````python
oa = df.groupby(["run", "episode"]).agg(
    steps=("steps", "max"),
    offtrack_rate=("on_track", lambda s: float((~s).mean())),
    crashes=("is_crashed", "sum"),
    max_objects=("n_objects", "max"),
    oa_enabled=("oa_enabled", "last"),
    status=("episode_status", "last"),
).reset_index()
display(oa.tail(10))
print("\nterminal episode_status counts:")
print(df.groupby("episode")["episode_status"].last().value_counts())
````
